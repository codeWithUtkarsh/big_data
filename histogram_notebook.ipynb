## Task 2.1 Draw the histogram with 'business-month' on the x-axis and trip counts on the y-axis. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data into a Pandas DataFrame for plotting - Task 2-1
trip_data = pd.read_csv("business_m_count.csv", sep=",", header=0)
trip_data.head
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Set Seaborn style
sns.set_style("whitegrid")

#logarithmic transformation to the 'count' column
trip_data['log_count'] = np.log10(trip_data['count'])

# Plot histogram for number of trips
plt.figure(figsize=(10, 6))
bars = sns.barplot(x="business_month", y="log_count", data=trip_data, palette="Blues")

# Add annotations for better understanding of the data
for bar in bars.patches:
    # Get the height of each bar
    height = bar.get_height()

    # Annotate the height of each bar on top of the bar
    plt.annotate(f'{height:.2e}',  # Format the height to scientific notation with two decimal places
                 xy=(bar.get_x() + bar.get_width() / 2, height),  # Position the annotation at the center of the bar
                 xytext=(0, 3),  # Offset the annotation by 3 points vertically
                 textcoords="offset points",  # Specify the text coordinates as offset points
                 ha='center', va='bottom',  # Align the annotation horizontally and vertically
                 fontsize=8, color='black')  # Adjust font size and color

# Set labels and title
plt.xlabel("Business Month", fontsize=12)
plt.ylabel("Trip Counts", fontsize=12)
plt.title("Number of Trips for Each Business in Each Month", fontsize=14)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Task 2.3 Draw the histogram with 'business-month' on the x-axis and the driver's earnings on the y-axis.

```python
# Load the earnings data into a Pandas DataFrame Task 2-3
earnings_data = pd.read_csv("business_m_driver_pay.csv", sep=",", header=0)

# logarithmic transformation to the 'count' column
earnings_data['log_val'] = np.log10(earnings_data['sum(driver_total_pay)'])

earnings_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the earnings data into a Pandas DataFrame
earnings_data = pd.read_csv("business_m_driver_pay.csv", sep=",", header=0)

# Plot a bar plot for driver's earnings
plt.figure(figsize=(10, 6))
bars = plt.bar(earnings_data["business_month"], earnings_data["sum(driver_total_pay)"], color='lightcoral')

# Add annotations to each bar for better understanding of the data
for bar in bars:
    # Get the height (earnings) of each bar
    height = bar.get_height()

    # Annotate the height of each bar on top of the bar
    plt.annotate(f'${height:,.2f}',  # Format the height to currency format with two decimal places and comma separator
                 xy=(bar.get_x() + bar.get_width() / 2, height),  # Position the annotation at the center of the bar
                 xytext=(0, 3),  # Offset the annotation by 3 points vertically
                 textcoords="offset points",  # Specify the text coordinates as offset points
                 ha='center', va='bottom',  # Align the annotation horizontally and vertically
                 fontsize=8, color='black')  # Adjust font size and color

# Set labels and title for better understanding
plt.xlabel("Business Month", fontsize=12)
plt.ylabel("Driver's Earnings ($)", fontsize=12)
plt.title("Monthly Earnings of Drivers", fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)  # Adjust font size of y-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for y-axis
plt.tight_layout()
plt.show()
```

## Task 5 - Draw the histogram with 'days' on the x-axis and ‘average waiting time’ on the y-axis

```python

# Load the average waiting time data into a Pandas DataFrame
average_waiting_time_pd = pd.read_csv("day_avg_req.csv", sep=",", header=0)

average_waiting_time_pd.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the average waiting time data into a Pandas DataFrame
average_waiting_time_pd = pd.read_csv("day_avg_req.csv", sep=",", header=0)

# Plot a bar plot for average waiting time in January
plt.figure(figsize=(10, 6))
plt.bar(average_waiting_time_pd["day"], average_waiting_time_pd["avg_request_to_pickup"], color='lightgreen')

# Set labels and title for better understanding
plt.xlabel("Days", fontsize=12)
plt.ylabel("Average Waiting Time (minutes)", fontsize=12)
plt.title("Average Waiting Time in January", fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)  # Adjust font size of y-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for y-axis
plt.tight_layout()
plt.show()
```
